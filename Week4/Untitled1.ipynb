# Using Boolean Masks for Finding Data

```python
theta = arange(0.,2*pi,pi/100)
plot(theta,sin(theta))
show()
```

```python
mask = (sin(theta)>0) & (sin(theta)<0.5)
plot(theta,sin(theta),theta[mask],sin(theta[mask]))
show()
#print mask
```

```python
y = sin(theta[mask])
jj = (y==max(y))
y[jj]=NaN
plot(theta,sin(theta))
plot(theta[mask],y,linewidth=4)
show()
```

# Interpolation

```python
theta = arange(0.,2*pi+0.01,pi/10)
mask = (theta==0.5)
#print mask
```

```python
newTheta = 0.5; newSin = 0.
from scipy.interpolate import interp1d
#help(interpld)
sinInterp = interp1d(theta,sin(theta))
newSin = sinInterp(newTheta)
print newSin
plot(theta,sin(theta),newTheta,newSin,"r+",markersize=21,markeredgewidth=3)
show()
```

```python
thetaNew=arange(0,2*pi+0.1,pi/30)
sinNew=interp1d(theta,sin(theta))(thetaNew)
plot(theta,sin(theta),thetaNew,sinNew,'r.')
show()
```

```python
plot(thetaNew,sinNew-sin(thetaNew))
ylabel('difference between interp and exact values')
show()
```

```python
sinNewC = interp1d(theta,sin(theta),kind='cubic')(thetaNew)
subplot(2,1,1)
plot(thetaNew,sinNew-sin(thetaNew),thetaNew,sinNewC-sin(thetaNew),'r')
subplot(2,1,2)
plot(thetaNew,sinNewC - sin(thetaNew),'r')
show()
```

# Differentiation

```python
x=arange(0.,10.01,0.1)
y=x**5
```

```python
dy=diff(y) #y(n+1)-y(n)
#help(diff)
print len(y), len(dy)
dx=diff(x)
dydx=dy/dx
print len(dydx)
```

```python
dydx1=zeros(y.shape,float)
dydx1[0:-1]=diff(y)/diff(x)
dydx1[-1] = (y[-1]-y[-2])/(x[-1]-x[-2])
```

$y=x^5$,$\frac{dy}{dx} = 5x^4$

```python
dydxExact=5*x**4
```

```python
plot(x,dydx1,x,dydxExact)
```
