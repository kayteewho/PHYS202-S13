# 3-D Plotting and Functions

\begin{equation}
V(x,y) = \frac{kq}{\sqrt{x^2+(y-d)^2}}- \frac{kq}{\sqrt{x^2+(y+d)^2}}
\end{equation}

```python
#make the xy-grid to cover the plane -2<x<2 and -2<y<2
dx = .06
dy = .06
x = arange(-2.,2.,dx)
y = arange(-2.,2.,dy)
print x.ndim, len(x)
X,Y = meshgrid(x,y)
print X.ndim
print Y.shape
print X
print Y
```

```python
#make the function, V(x,y)
k=1; q=1; d=1
V= (k*q/(X**2+(Y-d)**2)**(1/2.)) - (k*q/(X**2+(Y+d)**2)**(1/2.))
print V
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection ='3d')
p = ax.plot_surface(X, Y, V, rstride=1, cstride=1, linewidth=0, cmap=cm.YlGnBu)
# help(Axes3D.plot_surface)
cb = fig.colorbar(p, shrink = 0.5)
```

```python
#Represent 3-D data by contours
levels = array([8,4,2,1,0.5,0.25,0,-0.25,-0.5,-1,-2,-4,-8])

contour(X,Y,V,levels,linewidths=2,cmap=cm.Blues)
show()
```

# Exercise 1: Make a 3-D plot of the function $G(x,y) = e^{-x^2-y^2}$   over the range -6$\lt(x,y)\lt$6 using 51 bins in x and 51 bins in y.

```python
dx = .06
dy = .06
x = arange(-6.,6.,dx)
y = arange(-6.,6.,dy)
print x.ndim, len(x)
X,Y = meshgrid(x,y)
print X.ndim
print Y.shape
print X
print Y
```

```python
#make the function, G(x,y)
x=arange(-6,6,12./51.); y=arange(-6,6,12./51.); 
A,B = meshgrid(x,y)
G = exp(-A**2 - B**2)
print G 
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection ='3d')
p = ax.plot_surface(A, B, G, rstride=1, cstride=1, linewidth=0, cmap=cm.jet)
# help(Axes3D.plot_surface)
cb = fig.colorbar(p, shrink = 0.5)
```
